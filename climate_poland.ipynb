```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
temp = pd.read_excel('poland_temp_1901_2015.xls')
```

```python
rain = pd.read_excel('poland_rainfall_1901_2015.xls')
```

```python
rain = rain.drop([' ISO3', ' ISO2'], axis = 1)
```

```python
temp = temp.drop([' ISO3', ' ISO2'], axis = 1)
```

```python
rain.columns
```

```python
rain = rain.drop(' Country', axis = 1)
```

```python
temp = temp.drop(' Country', axis = 1)
```

```python
rain.columns = ['rainfall', 'year', 'month']
```

```python
temp.columns = ['temperature', 'year', 'month']
```

```python
climate = pd.merge(rain, temp, how = 'outer', on = ('year', 'month'))
```

```python
climate
```

```python
climate['day'] = 1
```

```python
climate['date'] = pd.to_datetime(climate[['year', 'month', 'day']])
```

```python
climate = climate.drop('day', axis = 1)
```

```python
climate
```

```python
climate.groupby('year')['temperature'].mean()
```
