```
import pandas as pd
import random
data = pd.DataFrame({'values':[random.random() for _ in range(10000)]})
data.head()
letters = ['A', 'B', 'C', 'D']
data['category'] = [random.choice(letters) for _ in range(data.shape[0])]
data = data.reindex(columns = ['category', 'values'])
data.groupby('category').agg({'values':'mean'}).reset_index()
import matplotlib.pyplot as plt
plt.plot(data['values'])
plt.show()
```
